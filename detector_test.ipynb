```python
from dataset.loader import GestureDataset
from solver.solver import Solver
from torchvision import transforms
import torch
import torch.nn as nn
import torch.optim as optim
from webcam_capturing.ResNetL import resnetl10
from torch.utils.data import DataLoader
import os
```

```python
path_annotations_train = os.path.join("D:", "IPN_Hand","annotations-20231128T085307Z-001", "annotations", "Annot_TrainList.txt")
path_annotations_test = os.path.join("D:", "IPN_Hand","annotations-20231128T085307Z-001", "annotations", "Annot_TestList.txt")
path_frames = os.path.join("D:", "IPN_Hand", "hand_gestures.h5")
```

```python
device = torch.device(f"cuda" if torch.cuda.is_available() else "cpu")

transform = transforms.Compose([
    transforms.ToTensor()
    ])

path_to_save = "here.pth"

ds_train = GestureDataset(path_frames, 
                            path_annotations_train, 
                            transform, 
                            sample_duration=60, 
                            label_all_gestures=True)

ds_test = GestureDataset(path_frames, 
                            path_annotations_test, 
                            transform, 
                            sample_duration=60, 
                            label_all_gestures=True)

model = resnetl10(num_classes = 2, sample_size = 128, sample_duration=60)

n_gesture = 3117
n_no_gesture = 4039 - 3117
weight = n_gesture / n_no_gesture
weight = torch.tensor([weight]).to(device)
criterion = nn.BCEWithLogitsLoss(pos_weight=weight)

optimizer = optim.Adam(model.parameters(), lr=0.001)

scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)

solver = Solver(model, 
                ds_train, 
                ds_test, 
                criterion, 
                optimizer, 
                scheduler, 
                device, 
                batch_size=32, 
                cnn_trans=False, 
                detector=True,
                save_every=5,
                path_to_save=path_to_save)
```
